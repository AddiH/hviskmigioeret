```python
import pandas as pd
import jiwer

from whisper_normalizer.basic import BasicTextNormalizer
normalizer = BasicTextNormalizer()

from evaluate import load
wer = load("wer")
```

```python
WER = pd.read_csv("../../data/WER.csv")
WER
```

```python
# in df WER, in index 19 of column folketinget, replace () with nothing 
print(WER["folketinget"][19])
WER["folketinget"][19] = WER["folketinget"][19].replace("(", "").replace(")", "")
print(WER["folketinget"][19])
```

```python
columns = WER.columns
columns_to_normalise = columns[1:7]

for column in columns_to_normalise:
    new_column_name = f"{column}_norm" 
    WER[new_column_name] = WER[column].apply(lambda x: normalizer(x))
```

```python
WER
```

```python
columns = WER.columns
columns_to_evaluate = columns[-6:]  # Adjust as needed
```

```python
for column in columns_to_evaluate:    
    # Find the index of empty strings in normalized columns
    empty_strings_indices = WER[WER[column] == ""].index

    # If there are any empty strings, print the corresponding original values
    if len(empty_strings_indices) > 0:
        original_column = column.replace("_norm", "")
        for idx in empty_strings_indices:
            print(f"Empty string found in column '{column}' at index {idx}.")
            print(f"Original value: {WER.at[idx, original_column]}")
    else:
        print(f"No empty strings found in column '{column}'.")
```

```python
columns_to_evaluate = columns[-5:]
# Function to compute WER for a single row for given columns
def compute_wer_for_row(row, column_name):
    reference_str = row['folketinget_norm']
    prediction_str = row[column_name]
    wer_score = wer.compute(predictions=[prediction_str], references=[reference_str])
    return round((wer_score*100),2)

# Iterate over each column that needs to be evaluated
for column in columns_to_evaluate:
    # Compute WER for each row and store the WER in a new column
    new_column_name = f"{column}_WER"  # Name for the new column holding WER scores
    
    # Use .loc to avoid SettingWithCopyWarning
    WER.loc[:, new_column_name] = WER.apply(lambda row: compute_wer_for_row(row, column), axis=1)
```

```python


# Dictionary to hold the actual string contents
strings_content = {
    "folketinget": folketinget,
    "folketinget_norm": folketinget_norm,
    "whisper_tiny": whisper_tiny,
    "whisper_small": whisper_small,
    "whisper_medium": whisper_medium,
    "whisper_large_v2": whisper_large_v2,
    "hviske": hviske,
    "whisper_tiny_norm": whisper_tiny_norm,
    "whisper_small_norm": whisper_small_norm,
    "whisper_medium_norm": whisper_medium_norm,
    "whisper_large_v2_norm": whisper_large_v2_norm,
    "hviske_norm": hviske_norm
}

# List of all string names for easy iteration
string_names = [
    "folketinget",
    "whisper_tiny", "whisper_small", "whisper_medium", "whisper_large_v2", "hviske",
    "folketinget_norm",
    "whisper_tiny_norm", "whisper_small_norm", "whisper_medium_norm", "whisper_large_v2_norm", "hviske_norm"
]

# Initialize a list to store the results
results = []

# Loop through each string name to calculate WER metrics
for string_name in string_names:
    hypothesis_text = strings_content[string_name]
    
    # Determine the appropriate reference string
    if string_name.endswith("_norm"):
        reference_text = strings_content["folketinget_norm"]
    else:
        reference_text = strings_content["folketinget"]
    
    # Handle the special case where the string is its own reference
    if string_name == "folketinget" or string_name == "folketinget_norm":
        s, d, i, h = 0, 0, 0, len(reference_text.split())
    else:
        out = jiwer.process_words(reference_text, hypothesis_text)
        s, d, i, h = out.substitutions, out.deletions, out.insertions, out.hits
    
    total_words = s + d + h
    
    # Append the results
    results.append({
        "String Name": string_name,
        "Hits": h,
        "Substitutions": s,
        "Deletions": d,
        "Insertions": i,
        "Total Words": total_words
    })

# Convert results to a DataFrame for display
import pandas as pd
results_df = pd.DataFrame(results)

# Display the DataFrame
results_df
```
