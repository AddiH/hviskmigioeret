```python
import pandas as pd
import jiwer

from whisper_normalizer.basic import BasicTextNormalizer
normalizer = BasicTextNormalizer()

from evaluate import load
wer = load("wer")
```

```python
WER = pd.read_csv("../../data/WER.csv")
WER
```

```python
# in df WER, in index 19 of column folketinget, replace () with nothing 
print(WER["folketinget"][19])
WER["folketinget"][19] = WER["folketinget"][19].replace("(", "").replace(")", "")
print(WER["folketinget"][19])
```

```python
columns = WER.columns
columns_to_normalise = columns[1:7]

norm = pd.DataFrame()

for column in columns_to_normalise:
    new_column_name = f"{column}_norm" 
    norm[new_column_name] = WER[column].apply(lambda x: normalizer(x))
```

```python
norm
```

```python
columns = norm.columns
columns_to_evaluate = columns[1:]  # Adjust as needed
columns_to_evaluate
```

```python
columns_to_evaluate = columns[1:]
# Function to compute WER for a single row for given columns
def compute_wer_for_row(row, column_name):
    reference_str = row['folketinget_norm']
    prediction_str = row[column_name]
    wer_score = wer.compute(predictions=[prediction_str], references=[reference_str])
    return round((wer_score*100),2)

# Iterate over each column that needs to be evaluated
for column in columns_to_evaluate:
    # Compute WER for each row and store the WER in a new column
    new_column_name = f"{column}_WER"  # Name for the new column holding WER scores
    
    # Use .loc to avoid SettingWithCopyWarning
    norm.loc[:, new_column_name] = norm.apply(lambda row: compute_wer_for_row(row, column), axis=1)
```

```python
norm
```

```python
import pandas as pd

# Assuming 'WER' is your DataFrame containing the WER columns
columns_wer = [
    'whisper-tiny_norm_WER',
    'whisper-small_norm_WER',
    'whisper-medium_norm_WER',
    'whisper-large-v2_norm_WER',
    'hviske_norm_WER'
]

# List to hold each model's WER statistics
stats_list = []

# Calculate statistics for each model
for column in columns_wer:
    min_wer = norm[column].min()
    max_wer = norm[column].max()
    mean_wer = norm[column].mean()
    # add median
    median_wer = norm[column].median()
    
    # Create a dictionary of the stats and add it to the list
    stats_dict = {
        'Model': column.replace('_WER', ''),  # Clean up the model name
        'Min WER': min_wer,
        'Max WER': max_wer,
        'Mean WER': round(mean_wer,2),
        'Median WER': median_wer
    }
    stats_list.append(stats_dict)

# Create a DataFrame from the list of dictionaries
norm_summary_matrix = pd.DataFrame(stats_list)

# Display the summary matrix
print(norm_summary_matrix)
```

```python
folketinget_norm = ' '.join(norm['folketinget_norm'].astype(str))
hviske_norm = ' '.join(norm['hviske_norm'].astype(str))
```

```python
hviske_norm
```

```python
wer_score = wer.compute(predictions=[hviske_norm], references=[folketinget_norm])
print(f"Hviske WER calculated for the entire string: {round((wer_score*100),2)}")
```

# testing string alignment

```python
import pandas as pd
import jiwer

# Assuming WER is your existing DataFrame
string_test = pd.DataFrame(norm, columns=['folketinget_norm', 'hviske_norm'])

# Initialize new columns
string_test['errors'] = 0
string_test['total_words'] = 0

# Define a function to process each row
def process_row(row):
    reference_string = row['folketinget_norm']
    hypothesis_string = row['hviske_norm']
    out = jiwer.process_words(reference_string, hypothesis_string)
    errors = out.substitutions + out.deletions + out.insertions
    total_words = out.substitutions + out.deletions + out.hits
    return errors, total_words

# Apply the function to each row
for index, row in string_test.iterrows():
    errors, total_words = process_row(row)
    string_test.at[index, 'errors'] = errors
    string_test.at[index, 'total_words'] = total_words

# add a column with the number of words in the reference string
string_test['total_words_reference'] = string_test['folketinget_norm'].str.split().str.len()
```

```python
print(f"Hviske norm WER calculated one string at a time: {round((sum(string_test['errors'])/sum(string_test['total_words']))*100,2)}")
print(f"Hviske WER calculated for the entire string: {round((wer_score*100),2)}")
```
